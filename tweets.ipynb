```python
#The following is a visualization of the first 1000 tweets pulled up directly from MongoDB. 

from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017/")
db = client['TwitterDB']
collection = db['tweets']

tweets = collection.find()

for tweet in tweets:
    print(f"_id: {tweet['_id']}")
    print(f"created_at: {tweet['created_at']}")
    print(f"id_str: {tweet['id_str']}")
    print(f"reply_count: {tweet['reply_count']}")
    print(f"retweet_count: {tweet['retweet_count']}")
    print(f"user: {tweet['user']}")
    print(f"user_name: {tweet['user_name']}")
    print(f"text: {tweet['text']}")
    print(f"hashtags: {tweet['hashtags']}")
    print(f"hashtag_count: {tweet['hashtag_count']}")
    print(f"keyword_count: {tweet['keyword_count']}")
    print("-----------")
```
